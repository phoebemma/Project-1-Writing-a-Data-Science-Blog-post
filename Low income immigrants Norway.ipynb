```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df1= pd.read_csv("Low income immigrant parents from Africa.csv", sep = ";", skiprows = 1)
df1.head()
```

```python
df2= pd.read_csv("Low income Africa_Asia_EastEurope.csv", sep = ";", skiprows = 1)
df2.head()
```

```python
df3= pd.read_csv("Low_income_immigrant parents from Nordic Region.csv", sep = ";", skiprows = 1)
df3.head()
```

```python
df4= pd.read_csv("Low_income_Nordic_countries.csv", sep = ";", skiprows = 1)
df4.head()
```

```python
Df= [df1, df2, df3,df4]
```

```python
df = pd.concat(Df)
df=df.reset_index(drop=True)
df.head()
```

```python
list(np.unique(df.contents))
```

To reduce the dimensionality, removing the "contents" row seems logical

```python
df=df.drop("contents", axis = 1)
df.head()
```

```python
df= df.melt(id_vars = ["group", "age"], var_name = "Year", value_name = "Proportion")
df.head()
```

```python
df.Proportion.min()
```

```python
df.Proportion.max()
```

```python
fig, ax =plt.subplots(figsize= (20,10), dpi = 50)
sns.set_theme(style="darkgrid")

sns.lineplot(y= "Proportion", x= "Year", data = df, hue = "group", linewidth =5, style = "group", ci = None )

ax.set_title("Low income immigrants in Norway across the years", fontdict = {"fontsize":25})
ax.set_xlabel("Year", fontsize = 25)
ax.set_ylabel("Proportion", fontsize = 25)
ax.tick_params(labelsize = 20)
ax.legend(loc= "upper center", bbox_to_anchor=(0.5, -0.1), fontsize = 20)
ax.set_ylim(3, 33, auto = True)
ax.set_xlim( auto= None )
ax.grid(linewidth= 4, which = "minor")
```

Since we can visualize the proportion of low income earners among immigrants in Norway based on their origins. I will then have to train a model to predict one's income level based on the data provided.

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.multioutput import MultiOutputClassifier
```

```python
data_f = pd.concat(Df)
data_f.head()
```

I will now separate the data into two with the "group" and "age" as the feature and the proportions as outcomes

```python
group = data_f[["group",]]
proportion = data_f.drop(["contents", "group", "age"], axis = 1)
group.head()
```

```python
proportion.head()
```

# Training the model

```python
X_train, X_test, y_train, y_test = train_test_split(proportion, group, test_size=0.33, random_state=42)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
# Making predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

Calculate the accuracy

```python
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
print('The training accuracy is', train_accuracy)
print('The test accuracy is', test_accuracy)
```
